3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
окружности,	
эллипса,
гиперболы.


```python
# Окружности
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 4, 4


x = []
y1 = []
y2 = []
r = 1
for i in range(-100, 101):
    xx = i/100
    x.append(xx)
    y1.append(np.sqrt(r**2-xx**2))
    y2.append(-np.sqrt(r**2-xx**2))
    
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('X')
plt.ylabel('Y')
```

```python
# эллипса
x = []
y1 = []
y2 = []
a = 3
b = 2
for i in range(-300, 301):
    xx = i/100
    x.append(xx)
    y1.append(np.sqrt((1-xx**2/a**2)*b**2))
    y2.append(-np.sqrt((1-xx**2/a**2)*b**2))
    
plt.axis([-3.5, 3.5, -3.5, 3.5])
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('X')
plt.ylabel('Y')
```

```python
# гиперболы
y = []
x1 = []
x2 = []
a = 1
b = 1
for i in range(-300, 301):
    yy = i/100
    y.append(yy)
    x1.append(np.sqrt((1+yy**2/b**2)/a**2))
    x2.append(-np.sqrt((1+yy**2/b**2)/a**2))
    
plt.axis([-3.5, 3.5, -3.5, 3.5])
plt.plot(x1, y)
plt.plot(x2, y)
plt.xlabel('X')
plt.ylabel('Y')
```

1. Задание (в программе)
Нарисуйте график y=k*cos(x-a)+b:   для некоторых (2-3 различных) значений параметров k, a, b

```python
x = np.linspace(-2*np.pi, 3*np.pi, 201)

k, k1, k2 = np.linspace(2, 10, num = 3)
a, a1, a2 = np.linspace(11, 20, num = 3)
b, b1, b2 = np.linspace(21, 30, num = 3)

plt.figure(figsize = (10, 10))
plt.plot(x, k * np.cos(x - a) + b, label='k, a, b')
plt.plot(x, k1 * np.cos(x - a1) + b1, color='r', label='k1, a1, b1')
plt.plot(x, k2 * np.cos(x - a2) + b2, color='b', label='k2, a2, b2')

plt.show()
```
