# 2. Работа будет состоять с недокументированным API. Нужно ввести релевантный запрос на сайте https://www.delivery-club.ru/search
(а) из предложенных точек с помощью API найти долю (в %) с бесплатной и платной доставкой. Для каждой категории рассчитать среднюю минимальную стоимость заказа.
(б) для каждой из категорий из пункта (а) рассчитать долю (в %) магазинов и ресторанов

## A

```python
url ='https://api.delivery-club.ru/api1.2/vendors/search'
params={'latitude':'56.844953',
        'longitude':'35.910127',
        'query':'манты'}
repos = requests.get(url =url, params=params)
vendors = repos.json()['vendors']
repos.json()
```

```python
countFreeDelivery  = 0
countStoreFreeDelivery = 0
countStore = 0
minPriceDelivFree = 0
minPriceDeliv =0
total = repos.json()['found']
for vendor in vendors:
    if (vendor['delivery']['price']['value'] == 0):
        countFreeDelivery+=1
        minPriceDelivFree+=vendor['delivery']['minOrderPrice']['value']
        if(vendor['categoryId']==3) :
            countStoreFreeDelivery +=1
        else: 
            minPriceDeliv+=vendor['delivery']['minOrderPrice']['value']
            if(vendor['categoryId']==3) :
                countStore +=1

result = {'Процент с бесплатной доставкой': f'{round(countFreeDelivery/total * 100,2)} %',
          'Процент с платной доставкой': f'{round((total - countFreeDelivery)/total* 100,2)} %',
          'Средняя минимальная стоимость заказа с бесплатной доставкой': round(minPriceDelivFree/countFreeDelivery,2),
          'Средняя минимальная стоимость заказа с платной доставкой': round(minPriceDelivFree/(total-countFreeDelivery),2),
          'Процент магазинов с бесплатной доставкой': f'{round(countStoreFreeDelivery/countFreeDelivery * 100,2)} %',
          'Рестораны с бесплатной доставкой': f'{round((countFreeDelivery - countStoreFreeDelivery)/countFreeDelivery * 100,2)} %',
          'Процент магазинов с платной доставкой': f'{round(countStore/(total - countFreeDelivery) * 100,2)} %',
          'Рестораны с платной доставкой': f'{round(((total - countFreeDelivery) - countStore)/(total - countFreeDelivery)  * 100,2)} %'}
result
```
